```python
from __future__ import division

import warnings
warnings.filterwarnings('ignore')
%config InlineBackend.figure_format = 'retina'
%pylab inline

import numpy as np
import pandas as pd
import re
import pprint

import rdflib
from rdflib.graph import Graph, ConjunctiveGraph
from rdflib import Graph, URIRef, BNode, Literal
from rdflib import RDF
from rdflib import Namespace
from rdflib.namespace import OWL, RDF, RDFS, FOAF, XSD
from sys import platform
import slugify
if platform == "linux" or platform == "linux2":
    prefix = "/home/massaro/"
elif platform == "darwin":
    prefix = "/Users/emanuele/"
    
dbo = rdflib.Namespace('http://dbpedia.org/ontology/')
schema = rdflib.Namespace('http://schema.org/')
time = rdflib.Namespace('http://www.w3.org/2006/time#')
```

```python
g = Graph()
ontology = Namespace("http://www.semanticweb.org/emanuele/ontologies/2017/8/herusOntology/")
```

```python
g.add((schema.City, RDFS.subClassOf, schema.Place))
g.add((dbo.PopulatedPlace, RDFS.subClassOf, schema.Place))
g.add((dbo.Settlement, RDFS.subClassOf, dbo.PopulatedPlace))
g.add((dbo.Town, RDFS.subClassOf, dbo.Settlement))
g.add((dbo.Region, RDFS.subClassOf, dbo.PopulatedPlace))
g.add((dbo.AdministrativeRegion, RDFS.subClassOf, dbo.Region))
g.add((dbo.GovernmentalAdministrativeRegion, RDFS.subClassOf, dbo.AdministrativeRegion))
g.add((dbo.Municipality, RDFS.subClassOf, dbo.GovernmentalAdministrativeRegion))
g.add((dbo.Canton, RDFS.subClassOf, dbo.GovernmentalAdministrativeRegion))
g.add((dbo.postalCode, RDF.type, RDF.Property))
g.add((dbo.postalCode, RDF.type, OWL.DatatypeProperty))
g.add((dbo.zipCode, RDF.type, RDF.Property))
g.add((dbo.zipCode, RDF.type, OWL.DatatypeProperty))
g.add((dbo.zipCode, RDFS.subPropertyOf, OWL.postalCode))
g.add((dbo.zipCode, RDFS.domain, dbo.PopulatedPlace))
```

```python
years = [2012, 2013, 2014, 2015, 2016, 2017]
fName = 'Dropbox (MIT)/HERUS_Lab/Data/Switzerland/keyData/keyDataSwiss%d.csv'%years[0]
data = pd.read_csv(prefix + fName,index_col=None, sep = ' ', encoding='utf-8')
for j in range(0,len(data)):  
    nameC = data.name[j]
    nameC = nameC.replace(" ", "_")
    idS = URIRef(ontology[nameC])
    #g.add( (idS, RDF.type, urbanSystem))
    for k in data.columns[:1]:   
        kk = data[k][j]
        kk = Literal(kk) # passing a string
        g.add((idS, RDF.type, dbo.Town)) 
        #g.add((idS, RDF.type, schema.City)) 
        g.add((idS, RDF.type, dbo.Municipality)) 
        g.add((idS, FOAF.Name, kk))
    for k in data.columns[1:2]:
        kk = data[k][j]
        kk = Literal(kk)
        g.add((idS, dbo.Canton, kk))
    for k in data.columns[2:3]:
        kk = data[k][j]
        kk = Literal(kk)
        g.add((idS, dbo.zipCode, kk))
    for k in data.columns[3:4]:
        kk = data[k][j]
        kk = Literal(kk)
        g.add((idS, dbo.zipCode, kk))
    for k in data.columns[4:5]:
        kk = data[k][j]
        kk = Literal(kk)
        g.add((idS, dbo.zipCode, kk))
```

```python
g.serialize(destination=prefix+'Desktop/outputProva.owl', format='turtle')
```
